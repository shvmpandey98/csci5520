### Colorado Groundwater Data

```python
import dataretrieval.nwis as nwis
```

```python
# Muddy Creek Well, CO
# USGD URL: https://waterdata.usgs.gov/nwis/inventory?site_no=393902103554001

site = '393902103554001'


# get instantaneous values (iv)
df_gwl_muddy_creek = nwis.get_record(sites=site, service='dv', start='2016-05-01', end='2022-11-30').reset_index()
df_gwl_muddy_creek = df_gwl_muddy_creek[['datetime', '72019_Mean']]
df_gwl_muddy_creek.columns = ['datetime', 'gwl']

display(nwis.get_record(sites=site, service='site'))

display(df_gwl_muddy_creek.head())
```

```python
# Drought Well, Near Pueblo
# USGD URL: https://waterdata.usgs.gov/nwis/inventory?site_no=382323104200701

site = '382323104200701'


# get instantaneous values (iv)
df_gwl_pueblo = nwis.get_record(sites=site, service='dv', start='2016-05-01', end='2022-11-30').reset_index()
df_gwl_pueblo = df_gwl_pueblo[['datetime', '72019_Mean']]
df_gwl_pueblo.columns = ['datetime', 'gwl']

display(nwis.get_record(sites=site, service='site'))

display(df_gwl_pueblo.head())
```

```python
df_gwl_muddy_creek.to_parquet('data/df_gwl_muddy_creek.parquet')
df_gwl_pueblo.to_parquet('data/df_gwl_pueblo.parquet')
df_gwl_kit_carson.to_parquet('data/df_gwl_kit_carson.parquet')
```
